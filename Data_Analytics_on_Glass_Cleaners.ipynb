```python
#Applying mode on catagorical data (Title)
print(products['Title'].mode())
print(products['Brand Name'].mode())
print(products['RateText'].mode())
```

```python
products['Title'].fillna(str(products['Title'].mode().values[0]), inplace=True)
products['Brand Name'].fillna(str(products['Brand Name'].mode().values[0]), inplace=True)
```

```python
products.isnull().sum()
```

```python
#Def Impute Median
def impute_median(series):
  return series.fillna(series.median())
```

```python
products.Rate=products['Rate'].transform(impute_median)
```

```python
#counting the null values in each column
products.isnull().sum()
```

```python
products['Title'] = pd.to_numeric(products['Title'],errors='coerce')
products['RateText'] = pd.to_numeric(products['RateText'],errors='coerce') #converting catagorical data into numeric data for analysis
```

```python
products['RateText'] = products['RateText'].apply(lambda x: str(x).replace('out of 5 stars','') if ' out of 5 stars' in str(x) else str(x))
products['RateText'] = products['RateText'].apply(lambda x: float(x))
products.head(10)
```

```python
products.describe()
```

```python
#Finding out the mean of ratings
grp=products.groupby('Brand Name')
x = grp['Rate'].agg(np.mean)
y = grp['Title'].agg(np.mean)
z = grp['RateText'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(16,5))
plt.plot(x, 'ro', color='r')
plt.xticks(rotation=90)
plt.title('Most loved brands')
plt.xlabel('Title------>')
plt.ylabel('Rating------->')
plt.show()
```
